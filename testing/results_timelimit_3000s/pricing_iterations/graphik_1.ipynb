```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
name1= "PR_Ins4_E_F.csv"
data1=pd.read_csv(name1)
name2= "PR_Ins4_E_A.csv"
data2=pd.read_csv(name2)
data2
```

```python
name1= "PR_Ins4_E_F.csv"
data1=pd.read_csv(name1)
name2= "PR_Ins4_E_A.csv"
data2=pd.read_csv(name2)
time1=data1["Unnamed: 5"].to_numpy()[1:]
time2=data2["Unnamed: 5"].to_numpy()[1:]

diff = len(time2) - len(time1)
if len(time2) < len(time1):
    for i in range(abs(diff)):
        time2=np.append(time2, 0)
else:
    for i in range(abs(diff)):
        time1=np.append(time1, 0)



x= np.arange(len(time2)) + 1 

fig, ax = plt.subplots()
ax.set_xticks(x)
plt.bar(x - 0.2, time1, 0.4, label = 'Exakt Farkas Pricing', color="blue")
plt.bar(x + 0.2, time2, 0.4, label = 'Exakt künstliche Variablen', color="gold")

x1, y1 = [4, 4], [0, max(time2)]
plt.plot(x1, y1, marker = '' , color="lime")

plt.legend()
plt.ylabel("Rechenzeit in [s]")
plt.xlabel("Iteration")
plt.savefig("Ins4_E_F_E_A.png", dpi=400)

```
