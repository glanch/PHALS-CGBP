```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc, rcParams
rc('text', usetex=True)
rcParams.update({'figure.autolayout': True})
```

```python
import numpy as np

def fill_array(A, K):
    diff = K - len(A)
    if diff <= 0:
        return A[:K]
    else:
        zeros = np.zeros(diff, dtype=A.dtype)
        return np.concatenate((A, zeros))

```

```python
name1= "PR_Ins4_E_F.csv"
data1=pd.read_csv(name1)
name2= "PR_Ins4_E_A.csv"
data2=pd.read_csv(name2)
time1=data1["Unnamed: 5"].to_numpy()[0:]
time2=data2["Unnamed: 5"].to_numpy()[1:]

diff = len(time2) - len(time1)
if len(time2) < len(time1):
    for i in range(abs(diff)):
        time2=np.append(time2, 0)
else:
    for i in range(abs(diff)):
        time1=np.append(time1, 0)



x= np.arange(len(time2)) + 1 

fig, ax = plt.subplots()
fig.set_figwidth(5)
fig.set_figwidth(5)
ax.set_xticks(x)

plt.bar(x - 0.2, time1, 0.3, label = 'Exakt; Farkas-Pricing', color="blue")

x1, y1 = [5, 5], [0, max(time2)]
plt.axvline(x=5, label="Zulässigkeit erreicht", color="red", linestyle="dashed")
#plt.plot(x1, y1, marker = ' ' , color="lime", linewidth=2, label="Zulässigkeit erreicht")
plt.bar(x + 0.2, time2, 0.3, label = 'Exakt; Künstl. Var.', color="gold")

avg_1 = sum(time1) / 7
avg_4_e_f = avg_1
avg_2 = sum(time2) / 7
avg_4_e_a= avg_2
plt.axhline(y=avg_1, label=r'$\emptyset$ Exakt; Farkas-Pricing')
plt.axhline(y=avg_2, label="$\emptyset$ Exakt; Künstl. Var.", color="orange")


handles, labels = plt.gca().get_legend_handles_labels()
order = [3,1,4,2,0]
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order]) 


plt.ylabel("Durchschnittliche Dauer in [s]")
plt.xlabel("Pricing-Iteration")

plt.savefig("Ins4_E_F_E_A.pdf") #dpi=400)

sum(time2)
#len(time2)
```
